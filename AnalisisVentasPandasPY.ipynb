```python
import pandas as pd
```

```python

# Regi칩n Norte
df_norte = pd.DataFrame({
    "producto": ["Laptop", "Mouse", "Pantalla", "Teclado", None],
    "cantidad": [5, 10, 3, None, 2],
    "precio_unitario": [1200, 25, 300, 50, None]
})

# Regi칩n Sur
df_sur = pd.DataFrame({
    "producto": ["Laptop", "mouse", "teclado", "Pantalla", "Laptop"],
    "cantidad": [3, 5, None, 1, 1],
    "precio_unitario": [None, 25, 50, None, 1200]
})

# Regi칩n Este
df_este = pd.DataFrame({
    "producto": ["Monitor", "Mouse", "Teclado", None, "Pantalla"],
    "cantidad": [2, None, 4, 2, 1],
    "precio_unitario": [250, 25, None, 100, 300]
})


df_productos = pd.DataFrame({
    "producto": ["Laptop", "Mouse", "Teclado", "Pantalla", "Monitor"],
    "categoria": ["Tecnolog칤a", "Accesorios", "Accesorios", "Pantallas", "Pantallas"],
    "stock": [10, 25, 8, 5, 12]
})
```

```python
#Combinar bases
df_regiones= pd.concat([df_norte,df_este,df_sur], keys=["Norte","Este","Sur"])
```

```python
#Validacion de datos
print("游 Informaci칩n general:")
print(df_regiones.info())

print("\n游늺 Shape:", df_regiones.shape)

print("\n游댌 Valores nulos:")
print(df_regiones.isna().sum())

print("\n游늵 Estad칤sticas:")
print(df_regiones.describe())
```

```python
#Limpieza
df_regiones["cantidad"]=df_regiones["cantidad"].fillna(0)
df_regiones["precio_unitario"]=df_regiones["precio_unitario"].fillna(df_regiones["precio_unitario"].mean())
df_regiones["producto"]=df_regiones["producto"].str.lower().str.capitalize().fillna("Desconocido")

```

```python
#Validacion de datos
print("游 Informaci칩n general:")
print(df_regiones.info())

print("\n游늺 Shape:", df_regiones.shape)

print("\n游댌 Valores nulos:")
print(df_regiones.isna().sum())

print("\n游늵 Estad칤sticas:")
print(df_regiones.describe())

print("\n游댌 Cabecera:")
print(df_regiones.head(3))
```

```python
df_completo= pd.merge(df_regiones,df_productos, on="producto", how="left")
df_completo
```

```python
#Limpieza de df total
df_completo["categoria"]=df_completo["categoria"].fillna("Desconocido")
df_completo["stock"]=df_completo["stock"].fillna(0)
```

```python
df_completo["ingresos"]=df_completo["cantidad"]*df_completo["precio_unitario"]
df_completo
```

```python
categorias_totales=df_completo.groupby(["categoria"])["cantidad"].sum()
print("\nEl total por categoria:")
print(categorias_totales)

poco_stock=df_completo.groupby("producto")["stock"].min().sort_values(ascending=True)
print("\nInventario poco stock:")
print(poco_stock)
total_producto_region = df_completo.groupby([df_completo.index.get_level_values(0), "producto"])["ingresos"].sum()
print("Total de ingresos por productos-region ")
print(total_producto_region)
```
